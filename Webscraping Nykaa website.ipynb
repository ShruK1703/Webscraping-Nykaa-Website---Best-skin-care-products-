##### Importing libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
# URL of the Nykaa website
url = "https://www.nykaa.com/skin/c/8377"
```

```python
r = requests.get(url)
```

```python
r.status_code
```

```python
htmlCode = r.text
```

```python
soup = BeautifulSoup(htmlCode, 'html.parser')
```

```python
print(soup.prettify())
```

### Extracting Product Name

```python
productname = soup.find_all('div',{'class':'css-xrzmfa'})
for p in productname:
    print(p.text)
```

### Extracting Original Price

```python
originalprice =  soup.find_all('span',{'class':'css-17x46n5'})
for price in originalprice:
    print(price.text)
```

### Extracting Offer Price

```python
offerprice=soup.find_all('span',{'class':'css-111z9ua'})
for offerp in offerprice:
    print(offerp.text)
```

### Extracting Discount

```python
discount = soup.find_all('span',{'class':'css-cjd9an'})
for dis in discount:
    print(dis.text)
```

### Extracting Offers/Free Gifts

```python
offer = soup.find_all('p',{'class':'css-1kzcg63'})
for o in offer:
    print(o.text)
```

### Extracting Reviews

```python
reviews = soup.find_all('span',{'class':'css-1qbvrhp'})
for r in reviews:
    print(r.text)
```

## GETTING URL OF ALL THE PAGES OF SKIN CARE PRODUCTS

```python
for i in range(1,801):
    print('https://www.nykaa.com/skin/c/8377?page_no={}&sort=popularity&eq=desktop'.format(i))
```

```python
name= []
originalprice = []
offerprice = []
discount = []
offer = []
reviews=[]


for i in range(1,801):
    Url = 'https://www.nykaa.com/skin/c/8377?page_no={}&sort=popularity&eq=desktop'.format(i)

    all_pages = requests.get(Url)
    htmlCode = all_pages.text

    soup = BeautifulSoup(htmlCode, 'html.parser')
    
    
    for x in soup.find_all('div', attrs = {'class' : 'css-d5z3ro'}):
        
        #Productname
        product= x.find('div', attrs={'class':'css-xrzmfa'})
        if product is None:
            name.append(np.NaN)
        else:
            name.append(product.text)
          
        #reviews
        rating = x.find('span', attrs = {'class':'css-1qbvrhp'})
        if rating is None:
            reviews.append(np.NaN)
        else:
            reviews.append(rating.text)
            
         #originalprice        
        orgprice =  x.find('span', attrs ={'class':'css-17x46n5'})
        if orgprice is None:
            originalprice.append(np.NaN)
        else:
            originalprice.append(orgprice.text)
            
            
        #offerprice
        off_price= x.find('span',{'class':'css-111z9ua'})
        if off_price is None:
            offerprice.append(np.NaN)
        else:
            offerprice.append(off_price.text)
            
        #discount
        disct = x.find('span',{'class':'css-cjd9an'})
        if disct is None:
            discount.append(np.NaN)
        else:
            discount.append(disct.text)
            
        
        #freegift
        gift = x.find('p',{'class':'css-1kzcg63'})
        if gift is None:
            offer.append(np.NaN)
        else:
            offer.append(gift.text)
            

            
print(len(name))
print(len(originalprice))
print(len(offerprice))
print(len(discount))
print(len(offer))
print(len(reviews))          
```

```python
#saving data into csv file

df = pd.DataFrame({'Product Name': name, 'Reviews' : reviews, 'Original Price' : originalprice, 'Offer Price' : offerprice, 'Discount': discount, 'Free Gifts': offer})
print(df.shape)
```
